```python
from statsmodels.tsa.seasonal import seasonal_decompose 
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
```

```python
#decompose the time series into trend, seasonality and residuals 

decomposition = seasonal_decompose(sales_data['Revenue'], model='additivee', period=12)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
#plot the decomposition trend

plt.figure(figsize=(12,8))
plt.subplot(411)
plt.plot(sales_data['Revenue'], label='Original')
plt.legend(loc="best")
plt.subplot(412)
plt.plot(trend, label="Trend")
plt.legend(loc="best")
plt.subplot(413)
plt.plot(seasonal, label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
#functions perfomrming the agumented dickey fuller test for stationary 

def test_stationarity(timeseries):
    result = adfuller(timeseries)
    print('ADF statistics:', result[0])
    print('p-value:', result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print(f' {key}:{value}')
```

```python
#testing stationarity of the original time series 

print('stationarity test for revenue:')
test_stationarity(sales_data['Revenue']) 
```

```python
#non-statinarity 

revenue_diff = sales_data['Revenue'].diff().dropna()
```

```python
#test stationarty of differenced time series 

print('Stationarity test for differences revenue:')
test_stationarity(revenue_diff)
```

```python
#split the data into traing and testing sets 

train_data = sales_data.iloc[:-12]
test_data = sales_data.iloc[-12:]
```

```python
#buid the model 

model_revenue = ARIMA(train_data['Revenue'], order=(1, 1, 1))
fitted_model_revenue = model_revenue.fit()
```

```python
#forecasitn the future rvenue 

forecast_revenue = fitted_model_revenue.forecast(steps=len(test_data)) 
```

```python
#plot the actual vs forecast revenues 

plt.figure(figsize=(12,6))
plt.plot(train_data.index, train_data['Revenue'], label='Training Data')
plt.plot(test_data.index, test_data['Revenue'], label='Actual Revenue')
plt.title('Revenue forecasting using ARIMA Model')
plt.xlabel('Period')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

```python
#evaluting the forecast accurcy for revenue 

forecast_error_revenue = test_data['Revenue'] - forecast_revenue
mae_revenue = np.mean(np.abs(forecast_error_revenue))
mse_revenue = np.mean(forecast_error_revenue ** 2)
rmse_revenue = np.sqrt(mse_revenue) 
```

```python
train_data = sales_data.iloc[:-12]
test_data = sales_data.iloc[-12:]
```

```python
# Define and train ARIMA model
model = ARIMA(train_data['Revenue'], order=(1, 1, 1))  # Example order (p, d, q)
fitted_model = model.fit() 
```

```python
# Forecast future sales
forecast_steps = len(test_data)
forecast = fitted_model.forecast(steps=forecast_steps)
```

```python
# Create forecast dataframe with datetime index
forecast_index = pd.date_range(start=test_data.index[0], periods=forecast_steps, freq='MS')
forecast_df = pd.DataFrame(data=forecast, index=forecast_index, columns=['Forecasted Revenue'])
```

```python
# Optional: Plot actual vs. forecasted sales
plt.figure(figsize=(12, 6))
plt.plot(train_data.index, train_data['Revenue'], label='Training Data')
plt.plot(test_data.index, test_data['Revenue'], label='Actual Sales')
plt.plot(forecast_df.index, forecast_df['Forecasted Revenue'], label='Forecasted Sales')
plt.title('Sales Forecasting using ARIMA Model')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

```python
# Calculate forecast error metrics (if needed)
forecast_error = test_data['Revenue'] - forecast
mae = np.mean(np.abs(forecast_error))
mse = np.mean(forecast_error ** 2)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error (MAE) for Forecast: {mae}')
print(f'Mean Squared Error (MSE) for Forecast: {mse}')
print(f'Root Mean Squared Error (RMSE) for Forecast: {rmse}')
```
